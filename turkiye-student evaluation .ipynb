```python
x=df.iloc[:,5:33]
from sklearn.decomposition import PCA
data=PCA(n_components=2,random_state=42)
x_data=data.fit_transform(x)
```

```python
x_data
```

```python
data.explained_variance_ratio_.cumsum()[1]
```

```python
from sklearn.cluster import KMeans
```

```python
distorsion=[]
for i in range(1,6):
    model=KMeans(n_clusters=i,n_jobs=-1,random_state=42)
    model.fit(x_data)
    distorsion.append(model.inertia_)
```

```python
distorsion
```

```python
plt.plot(range(1,6),distorsion)
plt.show()
```

```python
model=KMeans(n_clusters=3,n_jobs=-1,random_state=42)
model.fit(x_data)
y=model.predict(x_data)
```

```python
y
```

```python
cluster_=range(1,6)
plt.scatter(x_data[y==0,0],x_data[y==0,1],color='y',label='cluster 1')
plt.scatter(x_data[y==1,0],x_data[y==1,1],color='g',label='cluster 2')
plt.scatter(x_data[y==2,0],x_data[y==2,1],color='r',label='cluster 3')
plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1],s=100,color='b',label='centroid')
plt.title('cluster student')
plt.xlabel('pca 1')
plt.ylabel('pca 2')
plt.legend()
```
